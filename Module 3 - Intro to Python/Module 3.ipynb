### map function in pandas (but work in only one column at a time)

```python
raw_housing_data['Condition of the House'] = raw_housing_data['Condition of the House'].map({'Good':'1',
                                                                                            'Bad':'0',
                                                                                            'Fair':'1',
                                                                                            'Excellent':'3',
                                                                                            'Okay':'0'})
raw_housing_data['Condition of the House'].unique()
```

```python
raw_housing_data['Condition of the House']
```

```python
def year(value):
    return value.split()[-1]

raw_housing_data['Year Sold'] = raw_housing_data['Date House was Sold'].map(year)
raw_housing_data['Year Sold']
```

### Apply function

```python
# consider a home is luxury if at least house follows two of the criteria
def luxury_home(value):
    count = 0
    if value[0] == 'Yes':
        count = count + 1
    if value[1] in ['Good', 'Excellent']:
        count = count + 1
    if value[2] >= 8:
        count = count + 1
    if count >= 2:
        return 'Yes'
    else:
        return 'No'

raw_housing_data['Luxury Home'] = raw_housing_data[['Waterfront View', 'Condition of the House', 'Overall Grade']].apply(luxury_home, axis = 1)
raw_housing_data['Luxury Home'].unique()
```
